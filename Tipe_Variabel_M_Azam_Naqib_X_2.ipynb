#Assignment

```python
#Single Assignment
bb_fazel = 55
bb_dani = 60

print(bb_dani)
print(bb_fazel)
```

```python
bb_fazel, bb_dani = 50, 60, 70

print(bb_fazel)
print(bb_dani)
```

```python
#Multiple Assignment
bb_fazel, bb_dani = 55, 60

print(bb_dani)
print(bb_fazel)
```

```python

#jumlah variabel dengan nilai variabel harus sama banyak
bb_fazel, bb_dani = 55, 60, 70

print(bb_dani)
print(bb_fazel)
```

```python

#Chain Assignment
bb_fazel = bb_dani = 50

print(bb_fazel)
print(bb_dani)
```
